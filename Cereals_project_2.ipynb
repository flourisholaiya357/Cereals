```python
import pandas as pd
import numpy as np
```

```python
cereal_df = pd.read_csv("cereal.csv")
cereal_df.head(7)
```

```python
cereal_df.drop(['name'], axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
## LAbel Encoder helps in encoding the object type variables into binary(integers)

le = LabelEncoder()
cereals_df1 = cereal_df
cereals_df1.mfr = le.fit_transform(cereals_df1.mfr)
cereals_df1.type = le.fit_transform(cereals_df1.type)
cereals_df1
```

```python
cereals_df1.mfr
```

```python
y = cereals_df1.rating
X = cereals_df1.drop(['rating', 'name'], axis = 'columns')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
model_1 = LinearRegression()
model_1.fit(X_train, y_train)
model_1.score(X_test, y_test)
```

```python
## Mean Absolute Error shows the magnitude of difference between the prediction of the model and the true value of the data
from sklearn.metrics import mean_absolute_error
Y_preds = model_1.predict(X_test)
model_1_mae = mean_absolute_error(Y_preds, y_test)
print("MAE: ", model_1_mae)
```

```python
from sklearn.tree import DecisionTreeRegressor
model_2 = DecisionTreeRegressor()
model_2.fit(X_train, y_train)
model_2.score(X_test, y_test)
```

```python
model_2_predictions = model_1.predict(X_test)
model_2_mae = mean_absolute_error(y_test, model_2_predictions)
print("MAE: ", model_2_mae)
```

```python
from sklearn.ensemble import RandomForestRegressor
model_3 = RandomForestRegressor(n_estimators=100)
model_3.fit(X_train, y_train)
model_3.score(X_test, y_test)
```

```python
model_3_predictions = model_3.predict(X_test)
model_3_mae = mean_absolute_error(y_test, model_3_predictions)
print("MAE: ", model_3_mae)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
```

```python
## Pipeline Creation
## 1. Data preprocessing using standardscaler
## 2. Reduce dimension using PCA
## 3. Apply classifier
```

```python
pipeline_lr = Pipeline([('scaler_1', StandardScaler()),
                       ('pca_1', PCA(n_components=2)),
                       ('lr_classifier', LinearRegression())])
pipeline_dt = Pipeline([('scaler_2', StandardScaler()),
                       ('pca_2', PCA(n_components=2)),
                        ('dt_classifier', DecisionTreeRegressor())])
pipeline_rf = Pipeline([('scaler_3', StandardScaler()),
                       ('pca_3', PCA(n_components=2)),
                       ('rf_classifier', RandomForestRegressor())])

pipeline_model = (pipeline_lr, pipeline_dt, pipeline_rf)
```

```python
pipeline_dict = {0: "Linear Regression", 1: "DecisionTreeRegressor", 2: "Random Forest Regression"}
for pipe in pipeline_model:
    pipe.fit(X_train, y_train)
    
for i,model in enumerate(pipeline_model):
    print("{} test accuracy {}".format(pipeline_dict[i], model.score(X_test, y_test)))
```

```python
best_accuracy=0.0
best_classifier=0
best_pipeline=""

for i,model in enumerate(pipeline_model):
    if model.score(X_test, y_test)>best_accuracy:
        best_accuracy=model.score(X_test, y_test)
        best_pipeline=model
        best_classifier=i
        
print("Classifier with best accuracy: {} ".format(pipeline_dict[best_classifier]))
```
